ساخت Rule ها

```python
from mlxtend.frequent_patterns import association_rules
import pandas as pd
freq_01 = pd.read_pickle("out/freq_01.pkl")

rules = association_rules(
    freq_01,
    metric="lift",
    min_threshold=1.0
)

print("Total rules:", len(rules))
rules.sort_values("lift", ascending=False).head(10)
```

تمیز کردن جدول

```python
rules_clean = rules[[
    "antecedents", "consequents",
    "support", "confidence", "lift"
]].copy()

rules_clean = rules_clean.sort_values("lift", ascending=False).reset_index(drop=True)
rules_clean.head(10)
```

تبدیل product_id به اسم محصول

```python
# products باید از products.csv لود شده باشه
PRODUCTS_PATH = "products.csv"   # اگر تو مسیر دیگه‌ست، اصلاحش کن
products = pd.read_csv(PRODUCTS_PATH, usecols=["product_id", "product_name"])

print("Products loaded:", products.shape)
products.head()

id2name = dict(zip(products["product_id"], products["product_name"]))

def map_itemset(s):
    return [id2name.get(int(x), str(x)) for x in s]

top3 = rules_clean.head(3).copy()
top3["antecedents"] = top3["antecedents"].apply(map_itemset)
top3["consequents"]  = top3["consequents"].apply(map_itemset)
top3
```
