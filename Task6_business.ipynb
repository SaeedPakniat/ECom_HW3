## Task 6 – Business Insights

Using association rule mining on Instacart transaction data, several strong co-purchase patterns were discovered.

### Example Rule
- **Organic Raspberries → Organic Strawberries**
- Support: 0.013
- Confidence: 0.31
- Lift: 3.52

This rule indicates that customers who buy Organic Raspberries are more than three times as likely to also buy Organic Strawberries compared to random chance.

### Business Implications
- These products can be recommended together in a "Frequently Bought Together" system.
- They are good candidates for bundle promotions or cross-selling discounts.
- Product placement strategies can position these items closer in the app or warehouse.

Overall, association rules provide actionable insights that can increase basket size and improve customer experience.
